# 8.Split dependant and independant columns(x and y split)


```python
ak.head()
```

```python
y=ak.Age
y.head()
```

```python
X=ak.drop(columns=['Age'],axis=1)
X.head()
```

# 9.Data Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
```

```python
xsc=pd.DataFrame(sc.fit_transform(X),columns=X.columns)
xsc.head()
```

# 10.Train and Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(xsc,y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

# 11.Build the Model

```python
from sklearn.linear_model import LinearRegression 
model=LinearRegression()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_model=PolynomialFeatures(degree=6)
```

# 12.Train the Model

```python
model.fit(X_train,y_train)
```

```python
poly_x_values=poly_model.fit_transform(X)
```

```python
poly_model.fit(poly_x_values,y)
```

```python
reg_mod=LinearRegression()
reg_mod.fit(poly_x_values,y)
```

# 13.Test the ModelÂ¶

```python
pred=model.predict(X_test)
pred
```

```python
y_pred=reg_mod.predict(poly_x_values)
y_pred
```
